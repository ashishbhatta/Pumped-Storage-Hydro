**The df contain the required data between rows index 3 to 2980, so other data in head and tail part needs removing.**

```python
trimed_df = df.iloc[3:2980]
```

```python
trimed_df.head()
```

```python
trimed_df.tail()
```

Set First row of trimmed df as header

```python
new_header = trimed_df.iloc[0] #grab the first row for the header
trimed_df = trimed_df[1:] #take the data less the header row
trimed_df.columns = new_header #set the header row as the df header
trimed_df.head()
```

Remove Unwanted Columns

```python
trimed_df = trimed_df.drop(['Date', 'Time Block', 'Purchase Bid (MW)', 'Sell Bid (MW)', 'MCV (MW)',  'Final Scheduled Volume (MW)'],axis=1)
```

Converting Object datat type to integer and float

```python
trimed_df['Hour'] = trimed_df['Hour'].astype(int)
trimed_df['MCP (Rs/MWh) *'] = trimed_df['MCP (Rs/MWh) *'].astype(float)
print(trimed_df.dtypes)
```

Calculate Hourly average

```python
groupby = trimed_df.groupby('Hour')
data = groupby.mean()
data
```

List the smallest six values from the mean prices.

```python
data.nsmallest(6, 'MCP (Rs/MWh) *')
```

Determine the largest six prices from the mean values


```python
data.nlargest(6, 'MCP (Rs/MWh) *')
```
